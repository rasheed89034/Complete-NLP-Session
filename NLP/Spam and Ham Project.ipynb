## Load Dataset

```python
import pandas as pd 
```

```python
messages = pd.read_csv("SpamClassifier-master/smsspamcollection/SMSSpamcollection",sep='\t',names=['label','message'])
```

```python
messages.head()
```

```python
messages.tail()
```

```python
import nltk
import re 
from nltk.corpus import stopwords 
from nltk.stem import WordNetLemmatizer
```

```python
lemmetizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('^[a-zA-Z]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [lemmetizer.lemmatize(word,pos='v') for word in review if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Create Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,ngram_range = (1,2))
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
```

```python
np.set_printoptions(edgeitems=30,linewidth=100000,formatter=dict(float=lambda x:"%0.3g" % x))
X
```

```python
cv.vocabulary_
```

## Encoding for Output Feature(label)

```python
y = pd.get_dummies(messages['label'],dtype=int)
```

```python
y
```

```python
y = y.iloc[:,0].values
```

```python
y
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20,random_state=42)
```

## Train Your Model

```python
from sklearn.naive_bayes import MultinomialNB
spamDeduction = MultinomialNB()
```

```python
spamDeduction.fit(X_train,y_train)
```

```python
y_pred = spamDeduction.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
