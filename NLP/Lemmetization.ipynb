## Wordnet Lemmatizer
Lemmatization technique is like stemming. The output we will get after lemmatization is called ‘lemma’, which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma. Let us understand it with an example −

### Examples:
            Q&A, Chatbot, Text Summarization

'''
POS- Noun-n
verb-v
adjective-a
adverb-r
'''

```python
import nltk

nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("goes",pos='v')
```

```python
words = ["go","goes","gone","eat","eaten","eating","History","writing","writes","programming","programs"]
for word in words:
    print(word,"---->",lemmatizer.lemmatize(word,pos='v'))
```

```python
lemmatizer.lemmatize("fairly",pos='a'),lemmatizer.lemmatize("sportingly",pos='a')
```

```python
from nltk.tokenize import word_tokenize
```

```python
text = """In the realm of machine learning, the quality of your predictions is inherently tied to 
the quality of your data, making meticulous preprocessing an indispensable first step in any 
robust AI pipeline. Raw datasets are rarely ready for immediate analysis; they often contain inconsistencies, 
missing values, or unstructured text that can severely hinder an algorithm's performance. By diligently 
applying techniques such as feature scaling for numerical metrics or lemmatization for natural language, 
engineers transform chaotic information into a structured, digestible format. Ultimately, this rigorous data 
preparation is exactly what differentiates a fragile experimental model from a highly accurate, production-ready 
diagnostic tool."""
```

```python
words = word_tokenize(text)
print(words)
```

```python
for word in text:
    print(word_tokenize(word))
```

## Note:
For word tokenization we can use .split() fucntion also which don't required any library

```python
word_lemmatizer = WordNetLemmatizer()
lemmatizer_words = [word_lemmatizer.lemmatize(word,pos='v') for word in words]                      
```

```python
lemmatizer_words
```

## Used zip() 
for printing: Instead of print(words, "--->", lemmatizer_words) (which prints the giant lists over and over), zip(words, lemmatizer_words) smoothly loops through both lists at the exact same time, grabbing one word from each list per row.

```python
for original_word, lemma in zip(words, lemmatizer_words):
    print(f"{original_word} ---> {lemma}")
```
